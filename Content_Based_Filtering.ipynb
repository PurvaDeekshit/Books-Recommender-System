```python
fulldata = fulldata.loc[fulldata['isbn'].isin(all_isbn_list)]
```

```python
#Build Item profile
isbn_category = fulldata[['isbn','categories']]
isbn_category.shape
```

```python
isbn_category = isbn_category.loc[isbn_category['isbn'].isin(all_isbn_list)]
```

```python
isbn_category.drop_duplicates(inplace=True)
```

```python
print(isbn_category.shape)
print(isbn_category.head(10))
```

```python
isbn_category.isbn.unique()
```

```python
isbn_category.categories.unique()
```

```python
# one-hot encoding for category
uniq_categories = pd.get_dummies(isbn_category['categories'])
new_uniq_categories = uniq_categories.apply(lambda x: x/np.sqrt(uniq_categories.sum(axis=1)))
```

```python
isbn_category = pd.concat([isbn_category, new_uniq_categories], axis=1)
isbn_category.drop(columns='categories', inplace=True)
isbn_category.sort_values('isbn', inplace=True)
isbn_category.set_index('isbn', inplace=True)
```

```python
isbn_category.head(100)
```

```python
#Ratings for isbn
isbn_ratings = pd.pivot_table(fulldata, values='user_ratings', index=['isbn'], columns = ['user_id'])
isbn_ratings.sort_index(axis=1, inplace=True)
```

```python
print(isbn_ratings)
print(isbn_ratings.head(5))
```

```python
unique_users = isbn_ratings.columns
print(unique_users)
print(unique_users.value_counts().sum())
```

```python
#user to category mapping
user_category = pd.DataFrame(columns = isbn_category.columns)
print(user_category)
```

```python
for i in range(len(unique_users)):
    category_ratings = isbn_category.mul(isbn_ratings.iloc[:,i], axis=0)
    category_ratings.replace(0, np.NaN, inplace=True)    
    user_category.loc[unique_users[i]] = category_ratings.mean(axis=0)
```

```python
#user profile
user_category.head()
```

```python
#Calculate doucment and inverse document frequency
category_frequency = isbn_category.sum()
print(category_frequency)
inverse_document_frequency = (len(fulldata)/category_frequency).apply(np.log)
category_inv = isbn_category.mul(inverse_document_frequency.values)
```

```python
ratings_matrix = pd.DataFrame()
for i in range(len(unique_users)):
    newratings = category_inv.mul(user_category.iloc[i], axis=1)
    ratings_matrix[unique_users[i]] = newratings.sum(axis=1)
```

```python
ratings_matrix.to_pickle('content_based_rating_matrix.pkl')
```

```python
def contentbasedtop10(user):
    allbooks = ratings_matrix.index
    allratings = ratings_matrix[user]
    ratings_ranking = []
    for i in range(len(allbooks)):
        ratingslist=[]
        try:
            ratingslist.append(allratings[i])
            ratingslist.append(allbooks[i])
            ratings_ranking.append(ratingslist)
        except:
            ratingslist.append(int(0))
            ratingslist.append(allbooks[i])
            ratings_ranking.append(ratingslist)
    
    ratings_ranking = [item for item in ratings_ranking if item[1] in all_isbn_list]
    ratings_ranking.sort(key=lambda x: x[0])
    ratings_ranking.reverse()
    top10_ratings = ratings_ranking[:10]
    return top10_ratings
```

```python
#get isbn from recommended ratings
def top10isbn(recommendations):
    top10_isbn = []
    for item in recommendations:
        item_isbn = item[1]
        top10_isbn.append(item_isbn)
    return top10_isbn
```
