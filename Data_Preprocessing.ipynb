```python
books.drop(columns=['goodreads_rev_link','infoLink','maturityRating'],inplace=True)
```

```python
books.isnull().sum()
```

```python
#Get count of all books
count_books = books.isbn.count()
print("Total book count : ", count_books)
```

```python
#Get count of all unique books
unique_books = books.isbn.nunique()
print("No of unique books : ", unique_books)
```

```python
#Replace not available ratings with 0
books['averageRating'].fillna(0, inplace=True)
books['ratingsCount'].fillna(0, inplace=True)
books['goodreads_rating'].fillna(0, inplace=True)
books['goodreads_ratings_count'].fillna(0, inplace=True)
```

```python
books.isnull().sum()
```

```python
#Calculate weighted average from google ratings and good reads for each book
books['gr_rating_float'] = books['goodreads_rating'].astype(float) 
books['gr_rating_count_float'] = books['goodreads_ratings_count'].astype(float)
```

```python
books.shape
```

```python
books.head()
```

```python
#Calculate googleapi average*ratingcount
books['google_ratings'] = books.averageRating * books.ratingsCount
```

```python
#Calculate goodreads api average*ratingcount
books['goodreads_ratings'] = books.gr_rating_float * books.gr_rating_count_float
```

```python
#Calculate weighted average
books['weighted_avg'] = (books['google_ratings'] + books['goodreads_ratings'])/(books['ratingsCount'] + books['gr_rating_count_float'])
```

```python
books.dtypes
```

```python
books.shape
```

```python
books.head()
```

```python
#Keep only feature columns
columns_to_drop = ['averageRating','goodreads_rating','goodreads_ratings_count','industryIdentifiers','ratingsCount','google_ratings','gr_rating_float','gr_rating_count_float','goodreads_ratings']
books.drop(columns=columns_to_drop,inplace=True)
```

```python
books.head()
```

```python
books.drop(columns=['subtitle'],inplace=True)
```

```python
books.dropna(inplace=True)
```

```python
books.shape
```

```python
#Convert Authors from list to comma separated list
books['authors'] = books['authors'].apply(', '.join)
```

```python
books.head()
```

```python
#Convert Categories from list to comma separated values
books['categories'] = books['categories'].apply(', '.join)
```

```python
books.head()
```

```python
#Consider number of pages as integer
books['pageCount'] = books['pageCount'].astype(int)
```

```python
books.head()
```

```python
#Consider published date as year
books['publishedDate'] =  pd.to_datetime(books['publishedDate'])
books['publishedDate'] = pd.DatetimeIndex(books['publishedDate']).year
```

```python
books.head()
```

```python
#Rename publishedDate column to pubishedYear
books.rename(columns={'publishedDate':'publishedYear'}, inplace=True)
```

```python
books.head()
```

```python
#Split readingmodes json to multiple columns
reading_modes = pd.DataFrame(books['readingModes'].values.tolist())
reading_modes.columns = 'readingModes_'+ reading_modes.columns
reading_modes.columns
```

```python
books['readingModes_image'] = reading_modes[['readingModes_image']]
```

```python
books['readingModes_text'] = reading_modes[['readingModes_text']]
```

```python
#drop readingmodes column
books.drop(columns=['readingModes'],inplace=True)
```
